P3. ¿Cuál es la película con el mayor ratio ingreso por minuto? ¿Y el cortometraje? Dar nombres y ratios. Criterio: Película > 30min>= Cortometraje >= 5min.

```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
peliculas_original = pd.read_csv("/content/drive/MyDrive/Organización de Datos/TP1/Movie dataset/movies.csv")
```

```python
pelis = peliculas_original[ ["title", "budget", "runtime", "revenue"] ].dropna()
```

```python
largometrajes = pelis[pelis["runtime"] > 30]
largometrajes["Ingreso por minuto"] = largometrajes["revenue"] / largometrajes["runtime"]
maximo_ratio_largometrajes = largometrajes["Ingreso por minuto"].max()
largometrajes[largometrajes["Ingreso por minuto"] == maximo_ratio_largometrajes]
```

```python
cortometrajes = pelis[pelis["runtime"] <= 30]
cortometrajes = cortometrajes[cortometrajes["runtime"] >= 5]
cortometrajes["Ingreso por minuto"] = cortometrajes["revenue"] / cortometrajes["runtime"]
maximo_ratio_cortometrajes = cortometrajes["Ingreso por minuto"].max()
cortometrajes[cortometrajes["Ingreso por minuto"] == maximo_ratio_cortometrajes]
```
