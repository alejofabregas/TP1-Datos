C5. ¿Cúal es la película cuya entropía de las ratings que recibe es más alta? Esta será la película con opiniones más variadas. Considere sólo películas con más de 30 ratings.

```python
import pandas as pd
from scipy.stats import entropy
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
ratings_original = pd.read_csv("/content/drive/MyDrive/Organización de Datos/TP1/Movie dataset/ratings.csv")
peliculas_original = pd.read_csv("/content/drive/MyDrive/Organización de Datos/TP1/Movie dataset/movies.csv")
```

```python
pelis = peliculas_original[ ["id", "title", "vote_average", "vote_count"] ].dropna()

ratings = ratings_original.dropna()
#ratings = ratings.groupby("movieId").filter(lambda x : len(x) > 30)
ratings = ratings[ratings.groupby("movieId").movieId.transform("count")>30].copy() #De esta forma filtrar es mucho más rápido

ratings_por_id = ratings.groupby(["movieId"], sort=False)["rating"].apply(list).reset_index(name = "rating")
ratings_por_id["entropia"] = ratings_por_id["rating"].map(lambda x : entropy(x, base=2))
```

```python
max_entropia = ratings_por_id[ratings_por_id.entropia == ratings_por_id.entropia.max()]
id_max_entropia = max_entropia["movieId"].iloc[0]

peli_max_entropia = pelis.loc[pelis["id"] == id_max_entropia]
peli_max_entropia["title"].iloc[0]
```
