```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_original = pd.read_csv("/content/drive/MyDrive/Organización de Datos/TP1/Datasets/spotify.csv")
```

```python
#Distribución de reproducciones en el top 10 de Spotify Argentina

spotify_ar = spotify_original[spotify_original["Region"] == "ar"]
spotify_ar = spotify_ar[spotify_ar["Position"] <= 10]
spotify_ar

plt.figure(dpi=125)
sns.violinplot(data=spotify_ar, x="Position", y="Streams", color="#1DB954")
plt.xlabel("Posición", labelpad=15)
plt.ylabel("Reproducciones", labelpad=15)
plt.title("Distribución de reproducciones en el top 10 de Spotify Argentina", pad=15)
```

```python
#Bar plot de canciones con más semanas en el top 200 de spotify
spotify_global = spotify_original[spotify_original["Region"] == "global"]
numeros_1 = spotify_global[spotify_global["Position"] == 1]
numeros_1 = numeros_1.groupby("Track Name").size()
numeros_1 = numeros_1.to_frame().reset_index()
numeros_1.columns = ["Canción", "Días en el puesto número 1"]
numeros_1

plt.figure(dpi=150)
sns.barplot(data=numeros_1, x="Días en el puesto número 1", y="Canción", color="#1DB954")
plt.title("Días en el puesto número 1 del top 200 de Spotify", pad=15)
plt.xlabel("Días", labelpad=15)
plt.ylabel("")
```
